```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

#加载数据集
data=pd.read_csv('/Users/mac/Documents/2/Learning Analytics/SVM/students_data.csv')
data.head() #展示前5列数据
```

```python
#查看各个字段是否有缺失值

print(data.isnull().sum())
```

```python
#检查每个属性的特征类型

print(data.dtypes)
```

```python
#proprocess

x=data.drop('Class',axis=1)
y=data['Class']
x=pd.get_dummies(x)#把所有的非数值型变量转化为数值型
x.head()


```

```python
from sklearn.model_selection import train_test_split
#选取训练集和测试集
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=10)

```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

#选用Logistic Regression模型预测，并检验准确性

lr=LogisticRegression()
lr.fit(x_train,y_train)
predict_y=lr.predict(x_test)
print('predict',predict_y)
score=accuracy_score(y_test,predict_y)

print('Precision Rate：', score)
```

```python
#去掉变量SectionID之后再做Class与其他变量之间的Logistic回归

df=data
x=df.drop('Class',axis=1)
x=x.drop(['StageID', 'Semester','SectionID'],axis=1)

y=df['Class']
x=pd.get_dummies(x)
x.head(10)


x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=10)
#训练模型并检验准确率
lr=LogisticRegression()
lr.fit(x_train,y_train)
predict_y =lr.predict(x_test)
print('predict',predict_y)
score=accuracy_score(y_test,predict_y)
score
print('Precision Rate：', score)
```

```python
from sklearn import svm
#SVM
model = svm.SVC(kernel='linear', C=12)
model.fit(x_train,y_train)
model.score(x_test,y_test)
predicted = model.predict(x_test)

score1 =accuracy_score(y_test,predicted)
print(predicted)
print('SVM Precision Rate',score1)
```

```python
from sklearn.ensemble import RandomForestClassifier
#Random Forest
rfc_model = RandomForestClassifier()
rfc_model.fit(x_train, y_train)

predicted = rfc_model.predict(x_test)

score1 =accuracy_score(y_test,predicted)
print(predicted)

print('rfc Precision Rate',score1)
```

```python
from sklearn.neighbors import KNeighborsClassifier
#K-NN
model = KNeighborsClassifier(n_neighbors=7)
model.fit(x_train, y_train)#建立模型
predict_y = model.predict(x_test)

print('predict',predict_y)
score2 =accuracy_score(y_test,predict_y)
print('Precision Rate：', score2)
```
